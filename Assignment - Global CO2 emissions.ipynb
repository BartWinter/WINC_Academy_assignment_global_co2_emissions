```python
import pandas as pd
tables = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_carbon_dioxide_emissions')
countries = tables[1]
countries
```

```python
#Convert it into a non-Multilndex dataframe
countries.columns = ['country', 'co2_1990', 'co2_2005', 'co2_2017', '2017(% of world)', '2017 vs 1990: change (%)', 'per land area', 'per capita', 'total including lucf', 'total excluding lucf']
countries
```

```python
#Filter out the rows in this table that aren't countries 
countries = countries.loc[3:]
countries
```

```python
#Top 5 biggest CO2 producers based on the latest measurement
sorted_countries_2017 = countries.sort_values(by=['co2_2017'], ascending = False)
sorted_countries_2017
```

**Graph 1: CO2 of the bigger countries**

```python
import matplotlib.pyplot as plt
import numpy as np

#top5 = sorted_countries_2017.loc([42, 202, 67, 93, 158], [country, 1990, 2005, 2017])
top5 = sorted_countries_2017.iloc[1:6, 0:4] 

fix, ax = plt.subplots()
years = ["1990", "2005", "2017"]

for index, row in top5.iterrows():
  plt.plot(years, row[1:])
ax.set_title("The 5 biggest CO2 producers in the world", fontweight = "bold")
ax.set_xlabel("Year")
ax.set_ylabel("CO2 emission in Mt")
ax.legend(labels=["China", "United States", "European Union", "India", "Russia"],bbox_to_anchor=(1, 0.5) )
ax.grid(True)
plt.show

```

**Graph 2: worst and best changers**

```python
#Calculate the relative change in CO2 emission and adding new columns to the dataframe
countries["rel_1990"] = 100
countries["rel_2005"] = ((countries.loc[:,"co2_2005"]/countries.loc[:,"co2_1990"])*100)
countries["rel_2017"] = ((countries.loc[:,"co2_2017"]/countries.loc[:,"co2_1990"])*100)
countries
```

```python
#Sort values by rel_2017
sorted_countries_rel_2017 = countries.sort_values(by=["rel_2017"], ascending = False)
sorted_countries_rel_2017
```

```python
#Find the top three and bottom three countries with regards to lowering these emissions
sorted_countries_rel_2017 = sorted_countries_rel_2017.head(3).append(sorted_countries_rel_2017.tail(3))
sorted_countries_rel_2017
```

```python
#Select the right rows and columns voor the line chart

top_and_bottom_3 = sorted_countries_rel_2017.iloc[[0,1,2,3,4,5],[0,10,11,12]]
top_and_bottom_3
```

```python
#Plot the line chart

fig, ax = plt.subplots()

years = ["1990", "2005", "2017"]

for index, row in top_and_bottom_3.iterrows():
  plt.plot(years, row[1:])

ax.set_ylim((-10000,25000))
ax.set_title("Worst and best CO2 changers", fontweight = "bold")
ax.set_xlabel("Year")
ax.set_ylabel("Relative CO2 emission in Mt, compared to 1990")
ax.legend(['Greenland', 'Cambodja','Equatorial Guinea','Ukraine','Moldava','Armenia'],bbox_to_anchor=(1, 0.5))
ax.grid(True)
ax.minorticks_on()
fig.set_size_inches(8,6)
plt.show

```

```python
#Create a more useful chart: the same chart for all countries that had a least five MT of CO2 emissions in 1990

co2_1990_high = countries.loc[(countries.co2_1990 > 5)]
co2_1990_high
```

```python
#Sort values by rel_2017
sorted_countries_rel_2017_new = co2_1990_high.sort_values(by=["rel_2017"], ascending = False)
sorted_countries_rel_2017_new
```

```python
#Find the top three and bottom three countries with regards to lowering these emissions
sorted_countries_rel_2017_new = sorted_countries_rel_2017_new.head(3).append(sorted_countries_rel_2017_new.tail(3))
sorted_countries_rel_2017_new
```

```python
#Select the right rows and columns voor the line chart

top_and_bottom_3_new = sorted_countries_rel_2017_new.iloc[[0,1,2,3,4,5],[0,10,11,12]]
top_and_bottom_3_new
```

```python
#Plot the line chart

fig, ax = plt.subplots()

years = ["1990", "2005", "2017"]

for index, row in top_and_bottom_3_new.iterrows():
  plt.plot(years, row[1:], label = row[0])

ax.set_ylim((-100,1500))
ax.set_title("Worst and best CO2 changers (min CO2 emission in 1990 = 5 Mt)", fontweight = "bold")
ax.set_xlabel("Year")
ax.set_ylabel("Relative CO2 emission in Mt, compared to 1990")
ax.legend(['Greenland', 'Cambodja','Equatorial Guinea','Ukraine','Moldava','Armenia'],bbox_to_anchor=(1, 0.5))
ax.grid(True)
ax.minorticks_on()
fig.set_size_inches(8,6)
plt.show
```
